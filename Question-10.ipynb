```python
#Question-10

import nltk

# Function to count the number of verbs, nouns, pronouns, and adjectives in a given text
def count_pos(text):
    # Tokenize the text into sentences
    sentences = nltk.sent_tokenize(text)

    # Initialize counters
    verb_count = 0
    noun_count = 0
    pronoun_count = 0
    adjective_count = 0

    # Process each sentence
    for sentence in sentences:
        # Tokenize the sentence into words
        words = nltk.word_tokenize(sentence)
        # Perform part-of-speech tagging
        tagged_words = nltk.pos_tag(words)

        # Count the number of verbs, nouns, pronouns, and adjectives
        for word, tag in tagged_words:
            if tag.startswith('V'):  # Verb
                verb_count += 1
            elif tag.startswith('N'):  # Noun
                noun_count += 1
            elif tag == 'PRP':  # Pronoun
                pronoun_count += 1
            elif tag.startswith('J'):  # Adjective
                adjective_count += 1

    # Create a dictionary with the counts
    pos_counts = {
        'verbs': verb_count,
        'nouns': noun_count,
        'pronouns': pronoun_count,
        'adjectives': adjective_count
    }

    return pos_counts

# Main function
def main():
    text = "The quick brown fox jumps over the lazy dog. He loves to run in the park."
    pos_counts = count_pos(text)
    print("POS Counts:", pos_counts)

    # Additional Test Case 1
    text2 = "She is reading a book. The books on the shelf are old."
    pos_counts2 = count_pos(text2)
    print("POS Counts (Additional Test Case 1):", pos_counts2)

    # Additional Test Case 2
    text3 = "I saw a beautiful sunset at the beach. The waves were crashing on the shore."
    pos_counts3 = count_pos(text3)
    print("POS Counts (Additional Test Case 2):", pos_counts3)

# Call the main function
main()
```
